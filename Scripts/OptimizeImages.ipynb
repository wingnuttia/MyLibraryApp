```python
from PIL import Image
import os

# Set your input and output directories
input_folder = r"C:\Users\karen\Desktop\Book App\public\covers"
output_folder = r"C:\Users\karen\Desktop\Book App\public\covers2"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Max height for resized covers
MAX_HEIGHT = 400
QUALITY = 85  # JPEG quality

for filename in os.listdir(input_folder):
    if filename.lower().endswith(".jpg"):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename)

        try:
            with Image.open(input_path) as img:
                # Resize if height is too big
                if img.height > MAX_HEIGHT:
                    new_height = MAX_HEIGHT
                    new_width = int((new_height / img.height) * img.width)
                    img = img.resize((new_width, new_height), Image.LANCZOS)

                # Save with compression
                img.convert("RGB").save(output_path, "JPEG", quality=QUALITY, optimize=True)
                print(f"✅ Optimized: {filename}")
        except Exception as e:
            print(f"❌ Failed on {filename}: {e}")
```
